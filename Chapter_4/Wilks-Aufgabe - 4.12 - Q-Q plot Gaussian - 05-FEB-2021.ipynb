```python
# A3 Tempeartures June Climate Data for Guayquil, Ecuardor, 1951-1970

m1 = np.array([26.1,24.5,24.8,24.5,24.1,24.3,26.4,24.9,23.7,23.5,24.0,24.1,23.7,24.3,26.6,24.6,24.8,24.4,26.8,25.2]) # Canandaigua T max

print(m1.shape)
```

```python
obs = m1
z = (obs-np.mean(obs))/np.std(obs)

stats.probplot(z, dist="norm", plot=plt)
plt.title("Normal P-Q plot")
plt.show()
```

```python
x = np.linspace(23, 27, 20)
y = norm.pdf(x, np.mean(obs), np.std(obs))

plt.plot(x,y)
```
