```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
from scipy.stats import norm
import numpy as np
#x = np.random.randn(100)
data = [26.1,24.5,24.8,24.5,24.1,24.3,26.4,24.9,23.7,23.5,24,24.1,23.7,24.3,26.6,24.6,24.8,24.4,26.8,25.2]

# Aus https://im-coder.com/wie-kann-ich-fit-einer-gauss-kurve-in-python.html
mean,std=norm.fit(data)
#norm.fit versucht, passen Sie die Parameter einer Normalverteilung basierend auf den Daten. Und zwar in dem Beispiel oben mean etwa 2 und std ist etwa 5.

#Um die Handlung, die Sie tun können:
with plt.style.context('bmh'):
  
    fig, ax = plt.subplots(figsize=(12,4))
    ax.set(title="Histogram of Guyalua @ Ecuador June Temperature in °C")
    plt.hist(data, bins=5, normed=True)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    y = norm.pdf(x, mean, std)
    plt.plot(x, y)
    plt.legend(['Fit','Data'])
    # Beschriftung hinzufügen
    style = dict(size=10, color='gray')
    ax.text(25.5,0.5,r"$\mu: $"+f"{mean:.4}")
    ax.text(25.5,0.4,r"$\sigma: $"+f"{std:.2}")
    str()
    plt.show()
```

```python
data = [26.1,24.5,24.8,24.5,24.1,24.3,26.4,24.9,23.7,23.5,24,24.1,23.7,24.3,26.6,24.6,24.8,24.4,26.8,25.2]
data_fahrenheit=np.array(data)*9/5+32
print(data_fahrenheit)
# Aus https://im-coder.com/wie-kann-ich-fit-einer-gauss-kurve-in-python.html
mean,std=norm.fit(data_fahrenheit)
#norm.fit versucht, passen Sie die Parameter einer Normalverteilung basierend auf den Daten. Und zwar in dem Beispiel oben mean etwa 2 und std ist etwa 5.

#Um die Handlung, die Sie tun können:
with plt.style.context('bmh'):
  
    fig, ax = plt.subplots(figsize=(12,4))
    ax.set(title="Histogram of Guyalua @ Ecuador June Temperature in °F")
    plt.hist(data_fahrenheit, bins=5, normed=True)
    #xmin, xmax = plt.xlim()
    x = np.linspace(75, 80, 100)
    y = norm.pdf(x, mean, std)
    plt.plot(x, y)
    plt.legend(['Fit','Data'])
    # Beschriftung hinzufügen
    style = dict(size=10, color='gray')
    ax.text(25.5*9/5+32,0.25,r"$\mu: $"+f"{mean:.4}")
    ax.text(25.5*9/5+32,0.20,r"$\sigma: $"+f"{std:.2}")
    str()
    plt.show()
```
